## create distance matrix

```python
distance_matrix = pd.DataFrame(columns=['origin', 'destination', 'distance', 'duration'])

for i in range(0,len(ds_location_demand)):
    origin_postal_code = ds_location_demand.iloc[i]['Postal code'].replace(" ", "")
    for j in range(0, len(ds_location_demand)):
        current_index = len(distance_matrix)
        if (i != j):
            distance_matrix.loc[current_index, 'origin'] = ds_location_demand.iloc[i]['CFM Name']
            distance_matrix.loc[current_index, 'destination'] = ds_location_demand.iloc[j]['CFM Name']
            
            destination_postal_code = ds_location_demand.iloc[j]['Postal code'].replace(" ", "")
            current_url = 'https://maps.googleapis.com/maps/api/distancematrix/json?origins='+ origin_postal_code +'&destinations=' + destination_postal_code +'&key=AIzaSyBJziLN90S8r8v4Qe6aX7Beqr1NCuZT05U'
            print(current_url)
            with urllib.request.urlopen(current_url) as url:
                distance_data = json.loads(url.read().decode())
                distance_matrix.loc[current_index, 'distance'] = distance_data['rows'][0]['elements'][0]['distance']['value']
                distance_matrix.loc[current_index, 'duration'] = distance_data['rows'][0]['elements'][0]['duration']['value']
            
            
print('saving distance matrix to csv file...')
distance_matrix.to_csv(r'data/distance_matrix.csv')
print('done')
```

```python
distance_matrix_from_warehouse = pd.DataFrame(columns=['origin', 'destination', 'distance', 'duration', 'origin_postcode', 'destination_postcode'])

origin_postal_code = 'NE55HH' #ds_location_demand.iloc[i]['Postal code'].replace(" ", "")
for j in range(0, len(ds_location_demand)):
    current_index = len(distance_matrix_from_warehouse)
    distance_matrix_from_warehouse.loc[current_index, 'origin'] = 'Fareshare Warehouse'
    distance_matrix_from_warehouse.loc[current_index, 'destination'] = ds_location_demand.iloc[j]['CFM Name']
    distance_matrix_from_warehouse.loc[current_index, 'origin_postcode'] = 'NE55HH'
    distance_matrix_from_warehouse.loc[current_index, 'destination_postcode'] = ds_location_demand.iloc[j]['Postal code']

    destination_postal_code = ds_location_demand.iloc[j]['Postal code'].replace(" ", "")
    current_url = 'https://maps.googleapis.com/maps/api/distancematrix/json?origins=NE55HH&destinations=' + destination_postal_code +'&key=AIzaSyBJziLN90S8r8v4Qe6aX7Beqr1NCuZT05U'
    print(current_url)
    with urllib.request.urlopen(current_url) as url:
        distance_data = json.loads(url.read().decode())
        distance_matrix_from_warehouse.loc[current_index, 'distance'] = distance_data['rows'][0]['elements'][0]['distance']['value']
        distance_matrix_from_warehouse.loc[current_index, 'duration'] = distance_data['rows'][0]['elements'][0]['duration']['value']
            
            
print('saving distance matrix to csv file...')
distance_matrix_from_warehouse.to_csv(r'data/distance_matrix_from_warehouse.csv')
print('done')
```

## Retrieve Longitude and Latitude

```python
long_lat_df = pd.DataFrame(columns=['postal_code', 'longitude', 'latitude'])

for i in range(0,len(ds_location_demand)):
    postal_code = ds_location_demand.iloc[i]['Postal code'].replace(" ", "")
    current_url2 = 'https://maps.googleapis.com/maps/api/geocode/json?address='+ postal_code +'&key=AIzaSyBJziLN90S8r8v4Qe6aX7Beqr1NCuZT05U'
    print(str(i) + ' ---> ' + current_url2)
    with urllib.request.urlopen(current_url2) as url:
        location_data = json.loads(url.read().decode())
        long_lat_df.loc[i, 'postal_code'] = postal_code
        long_lat_df.loc[i, 'longitude'] = location_data['results'][0]['geometry']['location']['lng']
        long_lat_df.loc[i, 'latitude'] = location_data['results'][0]['geometry']['location']['lat']
            
            
print('saving location data to csv file...')
long_lat_df.to_csv(r'data/locations_data.csv')
print('done')
```

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler

coordinates = whiten(np.array(ds_location_data.iloc[:,2:4]))
db = DBSCAN(eps=0.25, min_samples=2).fit(coordinates)


core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

# #############################################################################
# Plot result
import matplotlib.pyplot as plt

# Black removed and is used for noise instead.
unique_labels = set(labels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)

    xy = coordinates[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=14)

    xy = coordinates[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()
```

```python
#ds_location_demand.head()
list_cluster_0 = []
list_cluster_1 = []
list_cluster_2 = []
list_cluster_3 = []
list_cluster_4 = []
list_cluster_5 = []
for e in range(0, 6):
    #print(ds_location_data.loc[ds_location_data['cluster'] == e])
    if(e == 0):
        list_cluster_0 = ds_location_demand.loc[ds_location_demand['cluster'] == e]['CFM Name'].tolist()
    elif(e == 1):
        list_cluster_1 = ds_location_demand.loc[ds_location_demand['cluster'] == e]['CFM Name'].tolist()
    elif(e == 2):
        list_cluster_2 = ds_location_demand.loc[ds_location_demand['cluster'] == e]['CFM Name'].tolist()
    elif(e == 3):
        list_cluster_3 = ds_location_demand.loc[ds_location_demand['cluster'] == e]['CFM Name'].tolist()
    elif(e == 4):
        list_cluster_4 = ds_location_demand.loc[ds_location_demand['cluster'] == e]['CFM Name'].tolist()
    elif(e == 5):
        list_cluster_5 = ds_location_demand.loc[ds_location_demand['cluster'] == e]['CFM Name'].tolist()

list_group_cluster = []
list_group_cluster.append(list_cluster_0)
list_group_cluster.append(list_cluster_1)
list_group_cluster.append(list_cluster_2)
list_group_cluster.append(list_cluster_3)
list_group_cluster.append(list_cluster_4)
list_group_cluster.append(list_cluster_5)
```

```python
def calculate_total_weight(food_van):
    #print(len(food_van))
    total_weight = 0
    for address_name in food_van:
        total_weight += float(ds_location_demand.loc[ds_location_demand['CFM Name'] == address_name, 'Demand*'])
    return float(total_weight)
```

```python
def calculate_total_distance(route_scenario):
    total_time = 0
    for route in route_scenario:
        for current_pos in range(0, len(route)-1):
            curr_series = ds_post_code_distance_matrix.loc[(ds_post_code_distance_matrix['origin'] == route[current_pos]) & (ds_post_code_distance_matrix['destination'] == route[current_pos+1]), 'distance']
#             print(route[current_pos] +'----' + route[current_pos+1] +'----' + str(curr_series.values))
            total_time += int(curr_series)
    return float(total_time)
```

```python
import time

best_route = 9999999
timeout = time.time() + 60*1   

while True:
    list_dispatch_plan = []
    current_dispatch = ['Fareshare Warehouse']
    for clusternumber in range(0, len(list_group_cluster)):
        random_cluster_member = random.sample(list_group_cluster[clusternumber], len(list_group_cluster[clusternumber]))
        for addressx in random_cluster_member:
            current_dispatch.append(addressx)
            current_weight = calculate_total_weight(current_dispatch) 
            if (int(current_weight) > 700):
                list_dispatch_plan.append(current_dispatch[:])
                current_dispatch.clear()
                current_dispatch = ['Fareshare Warehouse']
    
    if (calculate_total_distance(list_dispatch_plan) < best_route):
        best_route = calculate_total_distance(list_dispatch_plan)
        print('current best distance: ' + str(best_route))
    
    if time.time() > timeout:
        for delivery_no in range(0, len(list_dispatch_plan)):
            print('Delivery number: ' + str(delivery_no+1))
            print(list_dispatch_plan[delivery_no])
            print('##############################################')
        print("finish")
        break
        
```
